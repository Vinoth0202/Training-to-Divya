```python
import deepchem as dc

tasks, datasets, transformers = dc.molnet.load_delaney(featurizer='GraphConv')
train_dataset, valid_dataset, test_dataset = datasets
```

```python
model = dc.models.GraphConvModel(n_tasks=1, mode='regression', dropout=0.2,batch_normalize=False)
```

```python
model.fit(train_dataset, nb_epoch=100)
```

```python
metric = dc.metrics.Metric(dc.metrics.pearson_r2_score)
print("Training set score:", model.evaluate(train_dataset, [metric], transformers))
print("Test set score:", model.evaluate(test_dataset, [metric], transformers))
```

```python
solubilities = model.predict_on_batch(test_dataset.X[:10])
for molecule, solubility, test_solubility in zip(test_dataset.ids, solubilities, test_dataset.y):
    print(solubility, test_solubility, molecule)
```

For CF Dataset - IC50

```python
tasks, datasets, transformers = dc.molnet.load_bace_classification(featurizer='ECFP', split='stratified')
train_dataset, valid_dataset, test_dataset = datasets
#
```

```python
model = dc.models.GraphConvModel(n_tasks=1, mode='classification', dropout=0.2,batch_normalize=False)

model.fit(train_dataset, nb_epoch=50)
```

```python
metric = dc.metrics.Metric(dc.metrics.accuracy_score) # Using accuracy for classification
print("Training set score:", model.evaluate(train_dataset, [metric], transformers))
print("Test set score:", model.evaluate(test_dataset, [metric], transformers))
```

```python
predictions = model.predict_on_batch(test_dataset.X[:10])
print("\nPredictions on first 10 test samples:")
print(predictions)
print("Actual labels for first 10 test samples:")
print(test_dataset.y[:10])
```
