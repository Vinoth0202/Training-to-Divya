The below command is useful to read improper format of csv files

```python
import csv
import pandas as pd

with open("/content/CHEMBL1203870.csv", newline='') as f:
    dialect = csv.Sniffer().sniff(f.read(1024))
    f.seek(0)
    cftr = pd.read_csv(f, delimiter=dialect.delimiter)
cftr
```

```python
data = cftr.rename(columns={
    'Smiles': 'smiles',  # replace with actual column name
    'AlogP': 'target'   # replace with actual column name
})
data
```

```python
data[['smiles', 'target']].to_csv('formatted_data.csv', index=False)
```

```python
!chemprop train \
    --data-path formatted_data.csv \
    --save-dir model_output \
    --epochs 5 \
    --metric rmse

#  --dataset-type regression \
```

```python
import csv
import pandas as pd

with open("/content/Ic50.csv", newline='') as f:
    dialect = csv.Sniffer().sniff(f.read(1024))
    f.seek(0)
    ic50 = pd.read_csv(f, delimiter=dialect.delimiter)
ic50
```

```python
data = ic50.rename(columns={
    'Smiles': 'smiles',  # replace with actual column name
    'AlogP': 'target'   # replace with actual column name
})
```

```python
print(ic50.columns)
```

```python
print(data.columns)
```

```python
data[['smiles', 'target']].to_csv('formatted_data.csv', index=False)
```

```python
!chemprop train \
    --data-path formatted_data.csv \
    --save-dir model_output \
    --epochs 3 \
    --metric rmse

#  --dataset-type regression \
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
train_data[["smiles", "target"]].to_csv("train_data.csv", index=False)
test_data[["smiles", "target"]].to_csv("test_data.csv", index=False)
```
