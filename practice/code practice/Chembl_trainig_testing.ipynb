```python
import csv
import pandas as pd

with open("/content/Ic50.csv", newline='') as f:
    dialect = csv.Sniffer().sniff(f.read(1024))
    f.seek(0)
    ic50 = pd.read_csv(f, delimiter=dialect.delimiter)
ic50
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(ic50, test_size=0.2, random_state=42)
```

```python
train_data[["Smiles", "AlogP"]].to_csv("train_data.csv", index=False)
test_data[["Smiles", "AlogP"]].to_csv("test_data.csv", index=False)
```

```python
!chemprop train \
    --data-path train_data.csv \
    --save-dir train_output \
    --epochs 3 \
    --metric rmse
```

```python
!chemprop predict \
    --test-path test_data.csv \
    --model-path train_output  \
    --preds-path preds.csv

```
